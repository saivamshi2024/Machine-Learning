## **solving linear equations  using scipy.linalg.solver:**

```python
import numpy as np
from scipy.linalg import solve

#define the coefficient matrix A and the constant vector b
A = np.array([[2,3],[1, -1]])
b = np.array([7,1])

#solve the linear equation Ax =  b
x = solve(A, b)

print('solution:', x)
```

## **Finding Eigen values and Eigen values**

```python
from scipy.linalg import eig # define a square matrix
A = np.array([[4,5],[5,4]])

#find eigen values and eigen vectors
eigenvalues, eigenvectors = eig(A)

print('Eigenvalues:', eigenvalues)
print('Eigenvectors:',eigenvectors)

```

## **Numerical Integration using scipy.integratge.quad**

```python
from scipy.integrate import quad

# Define the integrand function
def integrand(x):
  return x**2

# Intergrate the fucntion from 0 to 1
result, error = quad(integrand, 0, 1)

print("Result of integration:", result)
print("Estimated error:", error)
```

## Optimization using scipy.optimize.minimize:


```python
from scipy.optimize import minimize

# Define the objective function
def objective(x):
  return x**2 + 5*x + 6

# minimize the objective function
result = minimize(objective, x0=0)

print("Minimum value:",result.fun)
print("Optimal solution:", result.x)
```

## Root finding using scipy.optimize.root_scalar

```python
from scipy.optimize import root_scalar

# Define the function whose root need to be found
def func(x):
  return x**2 - 4

  # Find the root of the function within the interval [1,3]
root = root_scalar(func, bracket = [1, 3])


print("Root of the function:",root.root)
```

## solving ordinary differential equations (ODEs) using scipy.integration.solve.ivp

```python
from scipy.integrate import solve_ivp

# Define the differential equation dy/dx = x+y
def dy_dx(x, y):
  return x + y

# Solve the ODE with initial condition y(0) =0
sol = solve_ivp(dy_dx,[0,1],[0])

# Print the solution
print("Solution of the ODE:", sol.y)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

**summary() : The summary() method provides a concise overview of statistical analysis results, particularly regression models, including coefficients, standard errors, p-values, relationship between the independent and dependent variables, and goodness-of-fit measures such as R-squared. It's a valuable tool for interpreting and assessing the significance of model parameters.**

```python
#Add a constant term to the independent variable x, creating a new array x1.
#This is done because the OLS method typically includes a constant term in the regression equation.
x1 = sm.add_constant(x)

# Fit a linear regression model to the data using OLS
results = sm.OLS(y, x1).fit()

#Print out a summary of the regression results
results.summary()
```

```python
dataset = pd.read_csv("/content/car_data.csv")
```

```python
dataset.head()
```

```python
#input
x = dataset.iloc[:, [2,3]].values
x
```

```python
#output
y = dataset.iloc[:, 4].values
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
knn = KNeighborsClassifier(n_neighbors = 7,leaf_size=30, metric='minkowski',p=2)
knn.fit(xtrain,ytrain)
```

```python
ypred = knn.predict(xtest)
```

```python
print("Predicted values:")
print(ypred)
```

```python
knn.score(xtest, ytest)
```

```python
print("\n\n Accuracy :", accuracy_score(ytest, ypred)*100)
```

```python
cm = confusion_matrix(ytest,ypred)
print("Confusion Matrix :\n", cm)
```

```python
fig,ax = plt.subplots(figsize = (5,5))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0,1), ticklabels = ('Predicted Does not Buy car', 'Predicted Buys car'))
ax.yaxis.set(ticks=(0,1), ticklabels = ('Actual Does not But car', 'Actual Buys car'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
  for j in range(2):
    ax.text(j, i, cm[i,j], ha='center', va='center', color = 'black')
plt.show()
```

```python
X_set = x
y_set = y
for i, j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
              c = ListedColormap(('red', 'green'))(i), label = j)
plt.xlim(0, 80)
plt.ylim(0, 180000)
plt.title('KNN(Training set)')
plt.xlabel('Age')
plt.ylabel('Annual Salary')
plt.legend()
plt.show()
```

```python
#Visualizing the test set
X_set = xtest
y_set = ytest
for i, j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
              c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Classifier(Test set)')
plt.xlabel('Age')
plt.ylabel('Annual Salary')
plt.legend()
plt.show()
```

# **Multinomial KNN**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
dataset = pd.read_csv("/content/iris.csv")
```

```python
dataset.head()
```

```python
#input
x = dataset.iloc[:, [0,1,2,3]].values
x
```

```python
#output
y = dataset.iloc[:, 4].values
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
knn = KNeighborsClassifier(n_neighbors = 7)
```

```python
knn.fit(xtrain,ytrain)
```

```python
ypred = knn.predict(xtest)
print("Predicted values:")
print(ypred)
```

```python
knn.score(xtest, ytest)
print("\n\n Accuracy :", accuracy_score(ytest, ypred)*100)
```

```python
cm = confusion_matrix(ytest,ypred)
print("Confusion Matrix :\n", cm)
```

```python
fig,ax = plt.subplots(figsize = (6,6))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0,1,2), ticklabels = ('Predicted Setosa', 'Predicted Versicolor','Predicted Virginica'))
ax.yaxis.set(ticks=(0,1,2), ticklabels = ('Actual Setosa', 'Actual Versicolor','Actual Virginica'))
ax.set_ylim(2.5, -0.5)
for i in range(3):
  for j in range(3):
    ax.text(j, i, cm[i,j], ha='center', va='center', color = 'white')
plt.show()
```

```python
neighbors = np.arange(1, 11)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

#loop over k vlaues
for i, k in enumerate(neighbors):
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(xtrain, ytrain)
  # Compute training and test data accuracy
  train_accuracy[i] = knn.score(xtrain, ytrain)
  test_accuracy[i] = knn.score(xtest, ytest)
# Generate plot
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Data Accuracy ')
plt.plot(neighbors, train_accuracy, label = 'Training Data Accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show
```

# **K Means Clustering**

# **21-06-2024**

```python
import matplotlib.pyplot as plt
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.cluster import KMeans
#turn the data into set of points
data = list(zip(x,y))
print(data)
inertias = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)
plt.plot(range(1,11), inertias, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data)
plt.scatter(x,y, c = kmeans.labels_)
plt.show()
```

## K means Algorithm

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import fcluster,linkage,dendrogram
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv("/content/iris.csv")
dataset.head()
```

```python
#input
x = dataset.iloc[:, [0,1,2,3]].values
x
```

```python
sns.pairplot(dataset)
```

```python
#finding the optimal number of clusters for k means classification
Elbow = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10,random_state = 0)
  kmeans.fit(x)
  Elbow.append(kmeans.inertia_)
#plotting the results onto a line graph allowing usto observe 'The Elbow'
plt.plot(range(1,11), Elbow, marker = 'o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Elbow')
#within cluster sum of squares
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10,random_state =0)
y_kmeans = kmeans.fit_predict(x)
```

```python
#visualizing the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
#plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Means')
plt.legend()
```

```python
distance_matrix = linkage(x, method = 'ward',metric = 'euclidean')
#create the dendrogram
dendrogram(distance_matrix)
#display the dendrogram
plt.show()
```

```python
dataset = pd.read_csv("/content/iris.csv")
dataset
```

```python
dataset.head
```

```python
x = dataset.iloc[:, [0,1,2,3]].values
x
```

```python
#Finding the optimum number of clusters for k-means classification
Elbow = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    Elbow.append(kmeans.inertia_)
#Plotting the results onto a Line graph, allowing us to observe ‘The Elbow’
plt.plot(range(1, 11), Elbow, marker='o')
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Elbow')
#within cluster sum of squares
plt.show()
```

```python
distance_matrix = linkage(x, method = 'ward', metric = 'euclidean')
# Create a dendrogram
dn = dendrogram(distance_matrix)
# Display the dendogram
plt.show()
```
